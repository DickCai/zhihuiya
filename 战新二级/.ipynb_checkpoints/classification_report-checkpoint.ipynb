```python
from sklearn.metrics import classification_report
import pandas as pd
import os
os.chdir('/Volumes/GT/智慧牙/战兴二级/战新二级result')
```

# Tier mapping

```python
tier1_mapping = pd.read_excel("tier_code_name_mapping.xlsx", sheet_name = 'tier1')
tier2_mapping = pd.read_excel("tier_code_name_mapping.xlsx", sheet_name = 'tier2')
tier_mapping = pd.concat([tier1_mapping, tier2_mapping], axis = 0)
tier_mapping
```

# data


```python
data=pd.read_csv('RESULT_tier2_prediction_suzhou.csv')
```

```python
data
```

# test

```python
test=pd.read_excel('苏州战新二级2000家测试集.xlsx').dropna()[['company_id','SEIC2']]
```

```python
test
```

```python
merged = pd.merge(data, test, on = 'company_id', how = 'inner')
```

```python
merged
```

```python
len(merged['SEIC2'].value_counts())
```

```python
ttl_names = list(set(merged['SEIC2'].drop_duplicates().tolist() + merged['tier2'].drop_duplicates().tolist()))
```

```python
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
```

```python
dict_target_names
```

```python
import warnings
warnings.filterwarnings('ignore')
merged['y_true'] = merged['SEIC2'].map(lambda x: dict_target_names[x])
merged['y_pred'] = merged['tier2'].map(lambda x: dict_target_names[x])
perform = pd.DataFrame(classification_report(merged['y_true'].tolist(), merged['y_pred'].tolist(), target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```

```python
rst_tier1 = pd.merge(data, tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'tier2', right_on = 'name', how = 'left')
rst_tier1
```

```python
rst_tier1['code_tier1'] = rst_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
rst_tier1
```

```python
rst_tier1 = pd.merge(rst_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
rst_tier1
```

```python
rst_tier1 = rst_tier1[['company_id', 'name_y']].rename({'name_y': 'tier1'}, axis = 1)
rst_tier1
```

```python
test_tier1 = pd.merge(test, tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'SEIC2', right_on = 'name', how = 'left')
test_tier1
```

```python
test_tier1['code_tier1'] = test_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
```

```python
test_tier1
```

```python
test_tier1 = pd.merge(test_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
test_tier1 = test_tier1[['company_id', 'name_y']].rename({'name_y': 'seic1'}, axis = 1)
test_tier1

merged_tier1 = pd.merge(rst_tier1, test_tier1, on = 'company_id', how = 'inner')

ttl_names = list(set(merged_tier1['seic1'].drop_duplicates().tolist() + merged_tier1['tier1'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
merged_tier1['y_true'] = merged_tier1['seic1'].map(lambda ele: dict_target_names[ele])
merged_tier1['y_pred'] = merged_tier1['tier1'].map(lambda ele: dict_target_names[ele])
perform_tier1 = pd.DataFrame(classification_report(merged_tier1['y_true'].tolist(), merged_tier1['y_pred'].tolist(), target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform_tier1.sort_values('f1-score', ascending = False)
```
