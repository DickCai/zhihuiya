# 对比

```python
Result_seic2=pd.read_excel('compare.xlsx')
Result_seic2=Result_seic2[['company_id','company_name_x','SEIC2']]
Result_seic2
```

```python
RESULT_seic2_qatest=pd.read_csv('RESULT_seic2_qatest2.csv')
RESULT_seic2_qatest
```

```python
RESULT_seic2_qatest.drop_duplicates(inplace=True)
```

```python
RESULT_seic2_qatest
```

```python
compare=pd.merge(RESULT_seic2_qatest,Result_seic2,how='left',on='company_id')
compare
```

```python
type(compare.loc[0]['SEIC2'])
```

```python
compare['change']=compare.apply(
    lambda ele: (ele['tier2'] if (ele['tier2'] in ele['SEIC2']) else ele['SEIC2'].split(",")[0].replace("[", "")) 
    if ("," in ele['SEIC2']) 
    else ele['SEIC2'],axis=1)
```

```python
tmp = compare.copy()
tmp['indi'] = compare.apply(
    lambda ele: (ele['tier2'] if (ele['tier2'] in ele['SEIC2']) else ele['SEIC2'].split(",")[0].replace("[", "")) 
    if "," in ele['SEIC2'] 
    else ele['SEIC2'],axis=1)
tmp[tmp['company_id'] == 'CBFBA960198075E56020CC3A2B8611B2']
```

```python
from sklearn.metrics import classification_report
report=classification_report(compare['change'], compare['tier2'])
print(report)
```
