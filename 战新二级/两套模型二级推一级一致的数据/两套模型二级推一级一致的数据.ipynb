# 世耀战新二级分类

```python
seic_lv2=pd.read_sql('select *  FROM company.china_company_seic_lv2_0923',con=engine)
```

```python
seic_lv2
```

```python
rst_tier1 = pd.merge(seic_lv2[['company_id','company_name','seic_lv2']], tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'seic_lv2', right_on = 'name', how = 'left')
rst_tier1['code_tier1'] = rst_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
rst_tier1 = pd.merge(rst_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
rst_tier1 = rst_tier1[['company_id', 'company_name','name_y']].rename({'name_y': 'tier1'}, axis = 1)
```

```python
rst_tier1.rename(columns={'tier1':'seic_lv1'},inplace=True)
rst_tier1
```

```python
rst_tier1.to_json('全国90万战新一级分类.json',orient='records',lines=True,force_ascii=False)
```

# 两套模型一致的战新一级分类

```python
sample_seic_lv1['change']=np.where(sample_seic_lv1['seic_lv1']==sample_seic_lv1['tier1'],1,0)
```

```python
sample_seic_lv1['change'].value_counts()
```

```python
sample_seic_lv1=sample_seic_lv1[sample_seic_lv1['change']==1]
sample_seic_lv1=sample_seic_lv1[['company_id','company_name','seic_lv1']]
sample_seic_lv1
```

```python
sample_seic_lv2=pd.merge(sample_seic_lv1,seic_lv2[['company_id','seic_lv2']],on='company_id')
sample_seic_lv2
```

```python
sample_seic_lv2.to_sql('sample_seic_lv2_1013',con=engine,index=False)
```
