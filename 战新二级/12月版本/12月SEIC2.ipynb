## SEIC2

```python
seic2=pd.read_excel(r'D:\数据\全国数据\战兴分类\战新二级分类\11月版本\Compare.xlsx')
seic2
```

```python
Compare=pd.merge(seic2[['company_id','company_name','SEIC2']],china_company_seic_lv2_1207_detail[['company_id','tier2_validated','seic_accuracy','description','business_scope']],how='left',on='company_id')
Compare
```

```python
Compare['change']=Compare.apply(
    lambda ele: (ele['tier2_validated'] if (ele['tier2_validated'] in ele['SEIC2']) else ele['SEIC2'].split(",")[0].replace("[", "")) 
    if ("," in ele['SEIC2']) 
    else ele['SEIC2'],axis=1)
```

```python
Compare['label']=np.where(Compare['tier2_validated']==Compare['change'],1,0)
```

```python
Compare
```

```python
Compare.to_excel(r'D:\数据\全国数据\战兴分类\战新二级分类\11月版本\Compare.xlsx',encoding='utf-8',index=False)
```

```python
ttl_names = list(set(Compare['change'].drop_duplicates().tolist() + Compare['tier2_validated'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
```

```python
merged_SEIC2=Compare
```

```python
import warnings
warnings.filterwarnings('ignore')
merged_SEIC2['y_true'] = merged_SEIC2['change'].map(lambda x: dict_target_names[x])
merged_SEIC2['y_pred'] = merged_SEIC2['tier2_validated'].map(lambda x: dict_target_names[x])
perform = pd.DataFrame(classification_report(merged_SEIC2['y_true'].tolist(), merged_SEIC2['y_pred'].tolist(), target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```

## SEIC1

```python
rst_tier1 = pd.merge(merged_SEIC2[['company_id','company_name','tier2_validated']], tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'tier2_validated', right_on = 'name', how = 'left')
rst_tier1['code_tier1'] = rst_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
rst_tier1 = pd.merge(rst_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
rst_tier1 = rst_tier1[['company_id', 'name_y']].rename({'name_y': 'tier1'}, axis = 1)
```

```python
test_tier1 = pd.merge(merged_SEIC2[['company_id','change']], tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'change', right_on = 'name', how = 'left')
test_tier1['code_tier1'] = test_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
test_tier1 = pd.merge(test_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
test_tier1 = test_tier1[['company_id', 'name_y']].rename({'name_y': 'seic1'}, axis = 1)
```

```python
merged_tier1 = pd.merge(rst_tier1, test_tier1, on = 'company_id', how = 'left')
ttl_names = list(set(merged_tier1['seic1'].drop_duplicates().tolist() + merged_tier1['tier1'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
merged_tier1['y_true'] = merged_tier1['seic1'].map(lambda ele: dict_target_names[ele])
merged_tier1['y_pred'] = merged_tier1['tier1'].map(lambda ele: dict_target_names[ele])
perform_tier1 = pd.DataFrame(classification_report(merged_tier1['y_true'].tolist(), merged_tier1['y_pred'].tolist(), target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform_tier1.sort_values('f1-score', ascending = False)
```
