```python
import pandas as pd
import os
os.chdir('D:\数据\全国数据\战兴分类\战新二级分类\innovation_company')
from sqlalchemy import create_engine
from sklearn.metrics import classification_report
import numpy as np
pd.set_option('max_rows',100)
engine=create_engine('mysql+pymysql://company:company@192.168.16.244/company',echo=False)
```

# tier_mapping

```python
tier_mapping=pd.read_excel(r'D:\数据\全国数据\战兴分类\战新二级分类\tier_mapping.xlsx')
tier_mapping
```

# 简介

```python
china_company_data=pd.read_sql('SELECT * FROM  company.china_company_data',con=engine)
china_company_data=china_company_data[['company_id','description','business_scope']]
china_company_data
```

# 全国数据战新二级分类 

```python
china_company_seic_lv2=pd.read_sql('select * FROM  company.china_company_seic_lv2',con=engine)
china_company_seic_lv2
```

```python
rst_tier1 = pd.merge(china_company_seic_lv2[['company_id','company_name','seic_lv2']], tier_mapping[['code_tier2', 'name']].dropna(), left_on = 'seic_lv2', right_on = 'name', how = 'left')
rst_tier1['code_tier1'] = rst_tier1['code_tier2'].map(lambda ele: int(str(ele).split(".")[0]))
rst_tier1 = pd.merge(rst_tier1, tier_mapping[['code_tier1', 'name']].dropna(), on = 'code_tier1', how = 'left')
rst_tier1 = rst_tier1[['company_id', 'name_y']].rename({'name_y': 'tier1'}, axis = 1)
```

```python
rst_tier1
```

```python
rst_tier1_desc=pd.merge(rst_tier1,china_company_data,on='company_id')
rst_tier1_desc
```

## innovation_company

```python
innovation_company=pd.read_csv('innovation_company_seic_lv1.csv')
```

```python
innovation_company
```

```python
result=pd.merge(rst_tier1_desc,innovation_company,on='company_id')
```

```python
result
```

```python
result['change']=np.where(result['tier1']==result['seic_lv1'],1,0)
```
