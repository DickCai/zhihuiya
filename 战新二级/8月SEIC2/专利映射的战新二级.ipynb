SELECT * FROM npd_st.caiqijun_test_company ctc
JOIN (SELECT company_id,ans_id FROM npd_dw.innovation_company)  ic
on ctc.company_id=ic.company_id
JOIN (SELECT ans_id,patent_id FROM patent_dw.patent_biblio_current_assignee) pbca
on ic.ans_id=pbca.ans_id
JOIN (SELECT * FROM patent_dw.patent_view_description_offset
WHERE api_key='patent_seic_level2_CN_v1')  pvdo
on pbca.patent_id=pvdo.patent_id

```python
import pandas as pd
import os
os.chdir('D:\数据\全国数据\战兴分类\战新二级分类\country_0818')
```

```python
patent_seic2=pd.read_csv('战新二级专利.csv')
patent_seic2
```

```python
patent_seic2.drop(columns=['company_id.1','ans_id.1','patent_id.1','chunk_size','update_ts','api_key','source','chunk_id'],inplace=True)
```

```python
patent_seic2.info()
```

```python
patent_seic2['api_data']=patent_seic2['api_data'].apply(lambda x:eval(x))
```

```python
patent_seic2['SEIC2']=patent_seic2['api_data'].apply(lambda x:x.get('predict_name'))
```

```python
patent_seic2=patent_seic2.explode('SEIC2')
patent_seic2
```

```python
patent_seic2['SEIC2'].value_counts()
```

```python
import numpy as np
def f(x):
    s = x.value_counts()
    return np.nan if len(s) == 1 else s.index[1]
```

```python
patent_seic2['code_2']= patent_seic2.groupby("company_name")["SEIC2"].transform(f)
```

```python
patent_seic2.drop_duplicates(subset=['company_name'],inplace=True)
```

```python
patent_seic2=patent_seic2[['company_id','company_name','ans_id','code_1','code_2']]
patent_seic2
```
