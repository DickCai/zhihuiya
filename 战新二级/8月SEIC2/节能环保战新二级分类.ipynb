# compare

## 全国数据战新二级分类  

```python
china_company_seic_lv2=pd.read_sql('select * FROM  company.china_company_seic_lv2',con=engine)
china_company_seic_lv2
```

## 节能环保名单 

```python
Energy=pd.read_excel('节能环保样本.xlsx',header=1)
Energy=Energy[['企业名称','战新行业二级']]
Energy
```

```python
Compare=pd.merge(Energy,china_company_seic_lv2,left_on='企业名称',right_on='company_name')
Compare
```

```python
ttl_names = list(set(Compare['战新行业二级'].drop_duplicates().tolist() + Compare['seic_lv2'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
```

```python
merged_SEIC2=Compare
```

```python
Compare.to_excel('Compare.xlsx',encoding='utf-8',index=False)
```

# SEIC2

```python
import warnings
warnings.filterwarnings('ignore')
merged_SEIC2['y_true'] = merged_SEIC2['战新行业二级'].map(lambda x: dict_target_names[x])
merged_SEIC2['y_pred'] = merged_SEIC2['seic_lv2'].map(lambda x: dict_target_names[x])
perform = pd.DataFrame(classification_report(merged_SEIC2['y_true'].tolist(), merged_SEIC2['y_pred'].tolist(), target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```
