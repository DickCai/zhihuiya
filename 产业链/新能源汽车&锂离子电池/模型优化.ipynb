```python
Graphene=pd.read_sql('select * from INDUSTRY_gkj_smx_20220908_company_info_v160',con=engine)
Graphene=Graphene.query('node_name=="光刻胶"')
Graphene
```

# Compare

```python
Compare=pd.merge(Graphene_test,Graphene[['company_name','node_id','node_name']],how='left',on=['company_name','node_id'],suffixes=['_true','_pre'])
Compare['predict']=Compare['node_name_pre'].apply(lambda x:1 if x=="光刻胶" else 0 )
perform = pd.DataFrame(classification_report(Compare['标注'].tolist(), Compare['predict'].tolist(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```
