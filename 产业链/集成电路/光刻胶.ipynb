# 光刻胶

```python
Photoresist=pd.read_csv('光刻胶待标注样本_matched.csv')
Photoresist
```

```python
if_title_matched=Photoresist.query('if_title_matched==1')
if_title_matched
```

```python
if_title_matched.drop_duplicates('company_name',inplace=True)
```

```python
result.drop_duplicates(subset=['company_name'],keep=False,inplace=True)
```

```python
result_all=pd.concat([result,if_title_matched])
result_all
```
