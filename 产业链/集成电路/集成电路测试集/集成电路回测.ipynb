## 集成电路产业链 

```python
node_name=pd.read_excel('/Users/caicaiqijun/智慧牙/数据/全国数据/产业链/集成电路产业链/文档/集成电路映射关系.xlsx',sheet_name='节点关系')
node_name
```

### 四级节点 

```python
node_name_4=IC_company[['company_id','company_name','node_name']]
node_name_4
```

### 三级节点 

```python
node_name_3=pd.merge(IC_company[['company_id','company_name','node_name']],
                      node_name[['三级节点','四级节点']],how='left',left_on='node_name',right_on='四级节点')
node_name_3.drop(['node_name','四级节点'],axis=1,inplace=True)
node_name_3.rename(columns={'三级节点':'node_name'},inplace=True)
node_name_3=node_name_3[node_name_3['node_name'].notnull()]
node_name_3
```

### 二级节点

```python
node_name_2=pd.merge(IC_company[['company_id','company_name','node_name']],
                      node_name[['二级节点','三级节点']],how='left',left_on='node_name',right_on='三级节点')
node_name_2.drop(['node_name','三级节点'],axis=1,inplace=True)
node_name_2.rename(columns={'二级节点':'node_name'},inplace=True)
node_name_2=node_name_2[node_name_2['node_name'].notnull()]
node_name_2
```

### 一级节点 

```python
node_name_1=pd.merge(IC_company[['company_id','company_name','node_name']],
                      node_name[['一级节点','二级节点']],how='left',left_on='node_name',right_on='二级节点')
node_name_1.drop(['node_name','二级节点'],axis=1,inplace=True)
node_name_1.rename(columns={'一级节点':'node_name'},inplace=True)
node_name_1=node_name_1[node_name_1['node_name'].notnull()]
node_name_1
```

```python
node_name_all=pd.concat([node_name_1,node_name_2,node_name_3,node_name_4])
node_name_all.drop_duplicates(subset=['company_id','node_name'],inplace=True)
node_name_all
```

### node_id 

```python
node_id=pd.read_excel('/Users/caicaiqijun/智慧牙/数据/全国数据/产业链/集成电路产业链/文档/集成电路映射关系.xlsx',sheet_name='节点名称')
node_id=node_id[['node_name','node_id']]
node_id
```

```python
node_name_all=pd.merge(node_name_all,node_id,on='node_name',how='left')
node_name_all
```

# Compare

## 节点精确率 

```python
Compare=pd.merge(IC_test_patent,node_name_all,how='left',on=['company_id','node_id'])
Compare['node_name'].fillna('其它',inplace=True)
```

```python
Compare=Compare.query('node_name_check!="其它"')
```

```python
Compare
```

# 测试集不包含其它

## 1.3  

```python
ttl_names = list(set(Compare['node_name_check'].drop_duplicates().tolist()+Compare['node_name'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
Compare['y_true'] = Compare['node_name_check'].map(lambda ele: dict_target_names[ele])
Compare['y_pred'] = Compare['node_name'].map(lambda ele: dict_target_names[ele])
#perform = pd.DataFrame(classification_report(Compare['y_true'].tolist(), Compare['y_pred'].tolist(),target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform = pd.DataFrame(classification_report(Compare['y_true'].tolist(), Compare['y_pred'].tolist(),target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```
