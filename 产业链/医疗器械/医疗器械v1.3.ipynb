# 测试集

```python
med_test=pd.read_excel('医疗器械精准率.xlsx',sheet_name='医疗器械')
med_test
```

# Compare

## 节点精确率 

```python
Compare=pd.merge(med_test,company[['company_id','node_id','node_name']],how='left',on=['company_id','node_id'],suffixes=['_true','_pre'])
Compare['node_name_pre'].fillna('other',inplace=True)
ttl_names = list(set(Compare['标注'].drop_duplicates().tolist()+Compare['node_name_pre'].drop_duplicates().tolist()))
dict_target_names = dict(zip(ttl_names, range(len(ttl_names))))
dict_target_names
Compare['y_true'] = Compare['标注'].map(lambda ele: dict_target_names[ele])
Compare['y_pred'] = Compare['node_name_pre'].map(lambda ele: dict_target_names[ele])
perform = pd.DataFrame(classification_report(Compare['y_true'].tolist(), Compare['y_pred'].tolist(),target_names=dict_target_names.keys(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```

## 产业链精确率

```python
med=pd.merge(med_test,company_one[['company_id','score']],on='company_id',how='left')
med
```

```python
med['label']=med['score'].apply(lambda x:1 if pd.isna(x)==False else 0)
```

```python
med
```

```python
perform = pd.DataFrame(classification_report(med['是否医疗器械相关'].tolist(), med['label'].tolist(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```
