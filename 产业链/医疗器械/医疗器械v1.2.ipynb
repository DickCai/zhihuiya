```python
patent=INDUSTRY_med_equip_patent_info_v120_1
patent
```

```python
patent['ans_id']=patent['ans_id'].str.extract(r'\[(.*)\]',expand=True)
patent
```

```python
company=pd.merge(med_company[['company_id','company_name','seic_lv2','node_name','ans_id']],
                patent[['ans_id','node_name','patent_id','apno']],on=['ans_id','node_name'])
company
```

```python
company_node=company.groupby(['node_name','company_name','seic_lv2']).agg({'apno':pd.Series.nunique}).reset_index().rename(columns={'score':'apno'})
company_node
```

```python
company_node.to_sql('INDUSTRY_med_Inter_company_info_v120',con=engine_chain,index=False,if_exists='replace')
```
