##  03-上市发债企业主营产品

```python
bonds=pd.read_excel('生物医药产业链样本.xlsx',sheet_name='03-上市发债企业主营产品')
```

```python
bonds_list=bonds[(bonds['主营产品披露名称'].str.contains('中药')==False) &
                 (bonds['主营标准产品名称'].str.contains('中药')==False)][['企业名称','主营产品明细']]
bonds_list.rename(columns={'主营产品明细':'简介'},inplace=True)
```

## 05-主题产业链关联企业表
