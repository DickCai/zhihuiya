# INDUSTRY_laser_20220915_patent_info_v160

```python
patent_info=pd.read_sql('select * from INDUSTRY_laser_20220915_patent_info_v160',con=engine)
patent_info
```

```python
patent_info['ans_id']=patent_info['ans_id'].str.extract(r'\[(.*)\]',expand=True)
```

```python
patent_info.drop_duplicates(subset=['node_name','ans_id','apno'],inplace=True)
```
