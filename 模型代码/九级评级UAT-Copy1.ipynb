### 降级企业统计（有过下降：包含波动）

```python
company=df_new.columns.tolist()
```

```python
def statistic(a,b,c,d,e,f,g,h,name1,name2,name3,name4,name5,name6,name7,name8):
    company1=[]
    company2=[]  
    company3=[] 
    company4=[] 
    company5=[] 
    company6=[] 
    company7=[] 
    company8=[] 
    for i in company:
        company1.append(df_new[i].tolist().count(a))
        company2.append(df_new[i].tolist().count(b))
        company3.append(df_new[i].tolist().count(c))
        company4.append(df_new[i].tolist().count(d))
        company5.append(df_new[i].tolist().count(e))
        company6.append(df_new[i].tolist().count(f))
        company7.append(df_new[i].tolist().count(g))
        company8.append(df_new[i].tolist().count(h))  
    pinji=pd.DataFrame()
    pinji[name1]=company1
    pinji[name2]=company2
    pinji[name3]=company3
    pinji[name4]=company4
    pinji[name5]=company5
    pinji[name6]=company6
    pinji[name7]=company7
    pinji[name8]=company8
    pinji.index=company
    return pinji
```
