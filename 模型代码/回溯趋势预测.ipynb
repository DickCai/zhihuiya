# 实际趋势

## history_backtrack

```python
history_backtrack=pd.read_csv('history_backtrack_actual.csv')
history_backtrack
```

```python
history_backtrack['actual']=(history_backtrack['score_20200601']+history_backtrack['score_20200701']+history_backtrack['score_20200801'])/3
```

```python
history_backtrack
```

## 20220601

```python
old_score=pd.read_csv('20220601.csv')
old_score
```

```python
old_score_trend=pd.merge(old_score,history_backtrack[['company_id','actual']],on='company_id')
old_score_trend
```

```python
import numpy as np
```

```python
old_score_trend['trend']=np.sign(old_score_trend.actual-old_score_trend.score_20200601)
```

```python
old_score_trend['trend'].value_counts()
```

```python
old_score_trend.query('trend==0')
```

```python
old_score_trend['trend']=np.sign(old_score_trend.actual>old_score_trend.score_20200601).astype(int)
```

```python
old_score_trend['trend'].value_counts()
```

# WLS预测趋势

## WLS_predict 

```python
WLS_predict=pd.read_parquet('WLS_predict')
WLS_predict_trend=pd.merge(WLS_predict,old_score,on='company_id')
WLS_predict_trend['trend']=np.sign(WLS_predict_trend.predict-WLS_predict_trend.score_20200601)
WLS_predict_trend['trend'].value_counts()
```

## Compare_WPL

```python
Compare=pd.merge(old_score_trend,WLS_predict_trend,on='company_id',suffixes=['_old','_pre'])
Compare
```

```python
perform = pd.DataFrame(classification_report(Compare['trend_old'].tolist(), Compare['trend_pre'].tolist(), output_dict = 'dict')).transpose()
perform.sort_values('f1-score', ascending = False)
```
