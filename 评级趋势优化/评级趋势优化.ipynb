# 合并

```python
rece_3_score=pd.merge(score_9,score_10,on='company_id',suffixes=['_9','_10'])
rece_3_score=pd.merge(rece_3_score,score_11,on='company_id')
rece_3_score
```

```python
rece_3_score['score_mean']=(rece_3_score['score_9']+rece_3_score['score_10']+rece_3_score['score'])/3
rece_3_score
```

```python
def f(x):
    if x>75.6270913121:
        rat='AAA'
    elif x>66.6745231748:
        rat='AA'
    elif x>56.5829515099:
        rat='A'
    elif x>45.4605119591:
        rat='BBB'
    elif x>33.7045067548:
        rat='BB'
    elif x>20.4460314446:
        rat='B'
    elif x>4.9992573616:
        rat='CCC'
    elif x>0.547162092:
        rat='CC'
    else:
        rat='C'
    return rat
```

```python
rece_3_score['rating_mean']=rece_3_score['score_mean'].apply(lambda x:f(x))
rece_3_score
```

```python
rece_3_score=pd.merge(rece_3_score,level_9,left_on='rating_mean',right_on='rating_v2')
rece_3_score
```
