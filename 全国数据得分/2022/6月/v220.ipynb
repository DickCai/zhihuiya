## 原始值对比 

```python
### 不同版本对比
def diff(cid,base,new):
    a=pd.concat([base.loc[cid],new.loc[cid]],axis=1)
    a.columns=['affter','front']
    for i in a.index:
        try:
            a.loc[i,'gap'] = float(a.loc[i,'affter'])-float(a.loc[i,'front'])
        except:
            a.loc[i,'gap'] = pd.NaT
    a.sort_values('gap',ascending=False,inplace=True)
    return a
```
