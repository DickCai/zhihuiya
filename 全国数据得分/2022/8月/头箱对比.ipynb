# gap

```python
Gap=pd.merge(original_top500_score[['company_id','company_name','seic_lv1','seic_lv2','total','rank_total','rank_seic_lv2','rating_seic_lv2']],
                test_quanguo_score[['company_id','total','rank_total','rank_seic_lv2','rating_seic_lv2']],
                on='company_id',suffixes=['_new','_old'])
Gap
```

```python
Gap['gap']=Gap['total_old']-Gap['total_new']
```

```python
Gap['gap'].min()
```

```python
pd.DataFrame(Gap.query('gap>0')['seic_lv2'].value_counts())
```

```python
pd.DataFrame(Gap.query('gap<0')['seic_lv2'].value_counts())
```
