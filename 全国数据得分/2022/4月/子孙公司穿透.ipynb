# 得分对比 

```python
def diff(cid,base,new):
    a=pd.concat([base.loc[cid],new.loc[cid]],axis=1)
    a.columns=['q3','q4']
    for i in a.index:
        try:
            a.loc[i,'gap'] = float(a.loc[i,'q3'])-float(a.loc[i,'q4'])
        except:
            a.loc[i,'gap'] = pd.NaT
    a.sort_values('gap',ascending=False,inplace=True)
    return a
```

```python
Compare=pd.merge(nationalgroup_totalscore,test_quanguo_score,on='company_id',suffixes=['_group','_com'])
```

```python
Compare['rank_gap']=abs(Compare['rank_total_group']-Compare['rank_total_com'])
Compare
```

```python
Company_100=Compare.query('rank_total_group>100 and rank_total_com<100')[['company_id','company_name_group','seic_lv1_group',
                                                             'seic_lv2_group','apno_cnt_group','apno_cnt_com','rank_total_group',
                                                             'rank_total_com']]
Company_100
```
