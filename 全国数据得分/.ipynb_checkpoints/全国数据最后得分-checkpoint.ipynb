```python
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']                          # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False 
os.chdir('D:\数据\全国数据\全国数据得分')
```

```python
data=[]
with open('v2_quanguo_df_formatted.json','r',encoding='utf-8') as file_input:
    for i in file_input:
        result=json.loads(i)
        company_id=result['company_id']
        company_name=result['企业名称']
        credit_code=result['统一社会信用代码']
        SEIC1=result['战略性新兴产业一层分类']
        region=result['企业所在区域']
        city=result['企业所在城市']
        address=result['企业注册地址']
        score=result['总分']
        level=result['等级']
        
        data.append([company_id,company_name,credit_code,SEIC1,region,city,address,score,level])
```

```python
df=pd.DataFrame(data,columns=['company_id','company_name','credit_code','SEIC1','region','city','address','score','level'])
```

# 行业分布

```python
class_sum=df['SEIC1'].value_counts().reset_index()
class_sum.rename(columns={'index':'SEIC1','SEIC1':'sum'},inplace=True)
```

```python
class_sum
```

```python
plt.figure(figsize=(20,10))
X=class_sum['SEIC1']
Y=class_sum['sum']
plt.bar(X,Y,width=0.8)
plt.xticks(rotation=90,fontsize=20)
for x,y in zip(X,Y):
     plt.text(x, y+0.05, '%.0f' % y, ha='center', va= 'bottom',fontsize=14)
plt.show()
```

# 评级分布

```python
level=df['level'].value_counts().reset_index()
level.rename(columns={'index':'level','level':'sum'},inplace=True)
```

```python
plt.figure(figsize=(20,10))
X=level['level']
Y=level['sum']
plt.bar(X,Y,width=0.8)
plt.xticks(rotation=90,fontsize=20)
for x,y in zip(X,Y):
     plt.text(x, y+0.05, '%.0f' % y, ha='center', va= 'bottom',fontsize=14)
plt.show()
```
