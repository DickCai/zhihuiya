## 比较

```python
def diff(cid,base,new):
    a=pd.concat([base.loc[cid],new.loc[cid]],axis=1)
    a.columns=['q3','q4']
    for i in a.index:
        try:
            a.loc[i,'gap'] = float(a.loc[i,'q3'])-float(a.loc[i,'q4'])
        except:
            a.loc[i,'gap'] = pd.NaT
    a.sort_values('gap',ascending=False,inplace=True)
    return a
```

```python
diff('青岛大荣伟业数控科技有限公司',patent,v)
```
