```python
test_quanguo_0906=pd.read_sql('SELECT * FROM test_quanguo_score',con=engine)
```

```python
test_quanguo_0906.head(5)
```

```python
test_quanguo_0906.info()
```

```python
test_quanguo_0906_rank_total=test_quanguo_0906[['company_id','qualification','patent','nonpatent','total','rank_total','rank_seic','apno_cnt']]
test_quanguo_0906_rank_total.head(5)
```

```python
v=pd.merge(test_quanguo_0906_rank_total,test_quanguo_0831,on='company_id',
          suffixes=('_new','_old'))
v.head(5)
```

```python
v_rank=v[['company_id','company_name','seic_lv1','total_new','rank_total_new','rank_seic_new','apno_cnt_new','total_old','rank_total_old',
          'rank_seic_old','apno_cnt_old']]
```

```python
v_rank.info()
```

```python
v_rank['seic_lv1'].value_counts()
```

```python
v_rank['gap']=abs(v['rank_total_new']-v['rank_total_old'])
```

```python
top100=v_rank.sort_values('gap',ascending=False).head(100)
```

```python
top100.to_excel('top100.xlsx',encoding='utf-8',index=False)
```
