# 指标差异

```python
def diff(cid,base,new):
    a=pd.concat([base.loc[cid],new.loc[cid]],axis=1)
    a.columns=['v1','v2']
    for i in a.index:
        try:
            a.loc[i,'gap'] = float(a.loc[i,'v1'])-float(a.loc[i,'v2'])
        except:
            a.loc[i,'gap'] = pd.NaT
    a.sort_values('gap',ascending=False,inplace=True)
    return a
```
